```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
```

```python

model = load_model('churn_model.h5')
with open('label_encoder.pkl', 'rb') as f:
    label_encoder = pickle.load(f)
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

```

```python


input_data = {
    "CreditScore": 600,
    "Geography": "France",
    "Gender": "     Female",
    "Age": 40,
    "Tenure": 3,
    "Balance": 60000,
    "NumOfProducts": 2,
    "HasCrCard": 1,
    "IsActiveMember": 1,
    "EstimatedSalary": 50000
}


df = pd.DataFrame([input_data])


df["Gender"] = df["Gender"].str.strip()
df["Geography"] = df["Geography"].str.strip()


df["Geography_France"] = 1 if df.loc[0, "Geography"] == "France" else 0
df["Geography_Germany"] = 1 if df.loc[0, "Geography"] == "Germany" else 0
df["Geography_Spain"] = 1 if df.loc[0, "Geography"] == "Spain" else 0


df.drop("Geography", axis=1, inplace=True)


```

```python
df["Gender"] = label_encoder.transform(df["Gender"])
```

```python
df
```

```python
scaler_df=scaler.transform(df)
```

```python
pred=model.predict(scaler_df)
```

```python
pred_pa=pred[0][0]
```

```python
pred_pa
```

```python
if pred_pa > 0.5:
    print("The customer is likely to churn.")
else:
    print("The customer is unlikely to churn.")
```
